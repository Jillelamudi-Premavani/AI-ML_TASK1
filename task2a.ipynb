```python
#Import Necessary Libraries:
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load Your Dataset:
# Assuming your dataset is in a CSV file
df = pd.read_csv('/content/Bank Customer Churn Prediction.csv')
```

```python
#Univariate Analysis - Numeric Variables
# Histogram for a numeric variable
plt.figure(figsize=(10, 6))
sns.histplot(df['credit_card'], bins=30, kde=True)
plt.title('Histogram of Credit card')
plt.show()
```

```python
df.head()
```

```python
# Boxplot for a numeric variable
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['age'])
plt.title('Boxplot of age')
plt.show()
```

```python
#Univariate Analysis - Categorical Variables:
# Countplot for a categorical variable
plt.figure(figsize=(10, 6))
sns.countplot(x='estimated_salary', data=df)
plt.title('Countplot of estimated_salary')
plt.show()
```

```python
#Univariate Analysis - Discrete Variables:
# Barplot for a discrete variable
plt.figure(figsize=(10, 6))
sns.barplot(x='tenure', y='tenure', data=df, estimator=sum)
plt.title('Barplot of Discrete Variable vs. Numeric Variable')
plt.show()
```

```python
#####bivariate aanalysis

#Scatter Plot for Two Numeric Variables:
# Scatter plot for two numeric variables
plt.figure(figsize=(10, 6))
sns.scatterplot(x='products_number', y='products_number', data=df)
plt.title('Scatter Plot of Numeric Variable1 vs. Numeric Variable2')
plt.show()
```

```python
#Correlation Matrix:
# Correlation matrix for numeric variables
correlation_matrix = df[['age', 'age', 'age']].corr()
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numeric Variables')
plt.show()
```
